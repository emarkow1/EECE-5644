## 2. Slicing with loc and iloc
`iloc` = by integer position, `loc` = by label.

```python
display(df.iloc[0])                                  # first row by position
display(df.loc[:, ["tenure", "MonthlyCharges"]].head())  # two columns by label
df2 = df.set_index("customerID")
df2.loc["7590-VHVEG"]                                 # one customer by ID
```

## 3. Filter rows by condition

```python
churned = df[df["Churn"] == "Yes"]
high_mtm = df[(df["Contract"] == "Month-to-month") & (df["MonthlyCharges"] > 80)]
print("churned:", len(churned), "| month-to-month & >$80:", len(high_mtm))
```

## 4. Sort, replace, rename

```python
df.sort_values("MonthlyCharges", ascending=False).head(3)
df["ChurnFlag"] = df["Churn"].replace({"Yes": 1, "No": 0})
df.rename(columns={"tenure": "MonthsWithCompany"}).head(2)
```

**Pro tip.** Avoid `inplace=True`. Treat DataFrames as immutable and assign the result to a new variable — it prevents subtle bugs in longer pipelines.

## 5. Descriptive stats and unique values
`value_counts` and `nunique` expose data-quality surprises before you encode.

```python
print(df["MonthlyCharges"].mean())
print(df["Contract"].value_counts())
print("distinct payment methods:", df["PaymentMethod"].nunique())
```

## 6. Handling missing values
In Telco, `TotalCharges` loaded as **text** because blank strings mark brand-new customers. Coerce to numeric to expose the gaps, then fill or drop.

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
print("missing TotalCharges:", df["TotalCharges"].isna().sum())
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
print("missing after fillna:", df["TotalCharges"].isna().sum())
```

*italicized text*## 8. Group and aggregate

```python
df["ChurnFlag"] = (df["Churn"] == "Yes").astype(int)
display(df.groupby("Contract")[["MonthlyCharges", "ChurnFlag"]].mean())
df.groupby("Contract").agg({"MonthlyCharges": "mean", "ChurnFlag": "sum"})
```

## 9. Apply a custom function

```python
def tenure_band(m):
    return "new" if m < 12 else ("mid" if m < 48 else "loyal")

df["TenureBand"] = df["tenure"].apply(tenure_band)
df.groupby("TenureBand")["ChurnFlag"].mean()
```

## 10. Combine DataFrames (concat / merge)

```python
a = df[["customerID", "tenure"]].head(3)
b = df[["customerID", "MonthlyCharges"]].head(3)
merged = pd.merge(a, b, on="customerID", how="left")   # SQL-style join
stacked = pd.concat([a, a], axis=0)                    # stack rows
print(merged.shape, stacked.shape)
merged
```
