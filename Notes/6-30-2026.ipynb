# Preprocessing

Making your data ready to be fed to a model. 

ML Life Cycle
1. Problem Definition
2. Data Collection
3. Data Preparation
4. Model Development
5. Model Evaluation
6. Deployment
7. Monitoring and Feedback

Six Step Toolkit
1. Import Libraries
    * Numpy
    * Pandas
    * Matplotlib
    * scikit-learn
2. Import dataset
    * split into X (inputs) and y (targets)
3. Handle missing data
    * Identify missing values
    * Impute using mean, median, or most-frequent value
4. Encode categoricals
5. Train/test split
6. Feature scaling

Data preprocessing takes 50-80% of the time in a project. Raw data is messy, everything downstream depends on having good data.
